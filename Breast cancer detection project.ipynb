# Feature Selection using SelectKBest METHOD

```python
#loading dataset
oversampled
```

```python
#importing libraries
from sklearn.feature_selection import SelectKBest, chi2, f_regression, f_classif

#specifying targeted coloumns 
predictors = ["radius_mean", "perimeter_mean", "area_mean", "compactness_mean", "concavity_mean", "concave points_mean", "radius_se", "perimeter_se", "area_se", "radius_worst", "perimeter_worst", "area_worst", "compactness_worst", "concavity_worst", "concave points_worst"]

#drop negative values
usethis = oversampled.dropna()
usethis.reset_index(drop=True)

usethis = usethis.filter(["radius_mean", "perimeter_mean", "area_mean", "compactness_mean", "concavity_mean", "concave points_mean", "radius_se", "perimeter_se", "area_se", "radius_worst", "perimeter_worst", "area_worst", "compactness_worst", "concavity_worst", "concave points_worst", "diagnosis"], axis=1)

usethisnow = usethis[(usethis["radius_mean"]>0) & (usethis["perimeter_mean"]>0) & (usethis["area_mean"]>0)
                     & (usethis["compactness_mean"]>0) & (usethis["concavity_mean"]>0) & (usethis["concave points_mean"]>0) 
                     & (usethis["radius_se"]>0) & (usethis["perimeter_se"]>0) & (usethis["area_se"]>0) & (usethis["radius_worst"]>0) 
                     & (usethis["perimeter_worst"]>0) & (usethis["area_worst"]>0) & (usethis["compactness_worst"]>0) & (usethis["concavity_worst"]>0) & (usethis["concave points_worst"]>0)]
usethisnow
#top N conditions are for target variable (i.e., diagnosis )
# Perform the feature selection
selector = SelectKBest(f_classif, k="all")
selector.fit(usethisnow[predictors], usethisnow["diagnosis"])

#get the p-values for each feature and then transform from p-values into some scores
scores = -np.log10(selector.pvalues_)

# Using plt to plot the scores!

plt.bar(range(len(predictors)), scores)
plt.xticks(range(len(predictors)), predictors, rotation='vertical')
plt.show()
```

# Machine Learning implementation

# Modelling

```python
# Importing libraries
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import svm
from matplotlib import pyplot
import numpy as np


#Selected features

#concave points_worst = 0.79
#perimeter_worst= 0.78
#radius_worst= 0.78
#concave points_mean = 0.77
#perimeter_mean = 0.75
#radius_mean = 0.74
#area_worst = 0.73
#concavity_mean = 0.69
#concavity_worst = 0.66
```

```python
#load dataset
oversampled
features = ["concave points_worst","perimeter_worst", "radius_worst", "concave points_mean", "perimeter_mean", "radius_mean", "area_worst", "concavity_mean", "concavity_worst"]
target = ['diagnosis']

#drop negative values
usethis = oversampled.dropna()
usethis.reset_index(drop=True)

usethis = usethis.filter(["concave points_worst", "perimeter_worst", "radius_worst", "concave points_mean", "perimeter_mean", "radius_mean", "area_worst", "concavity_mean", "concavity_worst", "diagnosis"], axis=1)

usethisnow = usethis[(usethis["concave points_worst"]>0) & (usethis["perimeter_worst"]>0) & (usethis["radius_worst"]>0)
                     & (usethis["concave points_mean"]>0) & (usethis["perimeter_mean"]>0) & (usethis["radius_mean"]>0)
                     & (usethis["area_worst"]>0) & (usethis["concavity_mean"]>0) & (usethis["concavity_worst"]>0)]

from sklearn.model_selection import train_test_split
y_target = usethisnow["diagnosis"].values
Xfeatures = usethisnow.drop("diagnosis", axis=1)

#The new bit is to use test-train split like so: 80:20

X_train, X_test, y_train, y_test = train_test_split(Xfeatures,y_target,test_size=0.2,random_state=1)   
```

# Feature Scaling

```python
#Feature Scaling of data
#oversampled dataset may contain features highly varying in magnitudes, range, and units. We need to bring all features to the same level of magnitudes. This can be done by Scaling the data, which means fitting the data within a specific range(example: 0-1).
#Using StandardScaler method in Scikit-Learn Library for scaling our data.
#importing library
from sklearn.preprocessing import StandardScaler 
ss = StandardScaler()
X_train = ss.fit_transform (X_train)
X_test = ss.fit_transform (X_test)
```

# Logistic Regression

```python
#fitting Logistic Regression to the training set 
lr = LogisticRegression(C=2.0, class_weight='balanced', dual=False, fit_intercept=True, intercept_scaling=1, l1_ratio=None, max_iter=20, multi_class='multinomial', n_jobs=None, penalty='l2', random_state=1, solver='newton-cg', tol=0.0001, verbose=0, warm_start=False)
lr.fit (X_train, y_train)
```

```python
#Predicting output with the test data
y_pred = lr.predict (X_test)
y_pred
```

```python
#Calculating test accuracy
#testing the model
y_pred

#importing accuracy score
from sklearn.metrics import accuracy_score

#printing the accuracy of the model
print(round(accuracy_score(y_test, y_pred) * 100, 0))
```

```python
#Training Time
%time history = lr.fit(X_train, y_train)
```

```python
#Confusion matrics
#A confusion matrix is a technique for summarizing the performance of a classification algorithm.
#Importing confusion matrics 
from sklearn.metrics import confusion_matrix
classifier=lr
conf_mat = confusion_matrix(y_test, classifier.predict(X_test))
sns.heatmap(conf_mat, square=True, annot=True, cmap='magma', fmt='d', cbar=True)
```

# Comparing Classifier Algorithms using RepeatedStratifiedKFold and cross_val_score

```python
#evaluate logistic regression model
lr= LogisticRegression(C=2.0, class_weight='balanced', dual=False, fit_intercept=True, intercept_scaling=1, l1_ratio=None, max_iter=20, multi_class='multinomial', n_jobs=None, penalty='l2', random_state=1, solver='newton-cg', tol=0.0001, verbose=0, warm_start=False)
cv1 = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores1 = cross_val_score(lr, Xfeatures, y_target, scoring='accuracy', cv=cv1, n_jobs=-1)
print(scores1)
print('LogisticRegression Mean Accuracy: %.3f (%.3f)' % (mean(scores1), std(scores1)))
```

```python
#evaluate decision tree classifier
dtc= DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=None, min_samples_split=5, min_samples_leaf=8, min_weight_fraction_leaf=0.0, max_features=None, random_state=1, max_leaf_nodes=None, min_impurity_decrease=0.0, class_weight=None, ccp_alpha=0.0)
cv2 = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores2 = cross_val_score(dtc, Xfeatures, y_target, scoring='accuracy', cv=cv1, n_jobs=-1)
print(scores2)
print('Decision tree Mean Accuracy: %.3f (%.3f)' % (mean(scores2), std(scores2)))
```

```python
#evaluate Random Forest classifier
rfc = RandomForestClassifier(n_estimators=104, criterion='gini', max_depth=None, min_samples_split=5, min_samples_leaf=3, min_weight_fraction_leaf=0.0, max_features='auto', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=True, n_jobs=-1, random_state=1, verbose=0, warm_start=False, class_weight=None, ccp_alpha=0.0, max_samples=None)
cv3= RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores3 = cross_val_score(rfc, Xfeatures, y_target, scoring='accuracy', cv=cv1, n_jobs=-1)
print(scores3)
print('Random Forest Mean Accuracy: %.3f (%.3f)' % (mean(scores3), std(scores3)))
```

```python
#evaluate Gradient boosting classifier 
gradient_booster = GradientBoostingClassifier(loss='log_loss', learning_rate=1.0, n_estimators=100, subsample=1.0, criterion='friedman_mse', min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_depth=3, min_impurity_decrease=0.0, init=None, random_state=1, max_features=None, verbose=0, max_leaf_nodes=None, warm_start=False, validation_fraction=0.1, n_iter_no_change=None, tol=0.0001, ccp_alpha=0.0)
cv4 = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores4 = cross_val_score(gradient_booster, Xfeatures, y_target, scoring='accuracy', cv=cv1, n_jobs=-1)
print(scores4)
print('Gradient boosting Mean Accuracy: %.3f (%.3f)' % (mean(scores4), std(scores4)))
```

# statistical hypothesis test

```python
#use a hypothesis test to see if the observed results are statistically significant.
#use the 5×2 procedure to evaluate the algorithms and calculate a p-value and test statistic value
from mlxtend.evaluate import paired_ttest_5x2cv
# check if difference between algorithms(Support vector machine and random forest classifier) is real
X=Xfeatures
y=y_target
t, p = paired_ttest_5x2cv(estimator1=rfc, estimator2=svc, X=X, y=y, scoring='accuracy', random_seed=1)
# summarize
print('P-value: %.3f, t-Statistic: %.3f' % (p, t))
```

```python
#interpret the p-value using an alpha of 5 percent.

# interpret the result
if p <= 0.05:
    print('Difference between mean performance is probably real')
else:
    print('Algorithms probably have the same performance')
    
#In this case, we can see that the p-value is about 0.589, which is much larger than 0.05. This leads us to fail to reject the null hypothesis, suggesting that any observed difference between the algorithms is probably not real.

#We could just as easily choose support vector machine or Random forest classifier and both would perform about the same on average.
```
